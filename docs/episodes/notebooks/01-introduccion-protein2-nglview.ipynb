## Step 1


```
import os
from pathlib import Path
import nglview as nv

course_dir = Path(os.environ.get('COURSE_DIR', '~/Concepcion26')).expanduser()
complex_pdb = course_dir / 'data' / 'complex' / 'protein2-ligand2.pdb'
view = nv.show_file(str(complex_pdb))
view.clear_representations()
view.add_cartoon(opacity=0.8)
view.add_ball_and_stick(selection='ligand', opacity=1.0)
view.center()
view
```
